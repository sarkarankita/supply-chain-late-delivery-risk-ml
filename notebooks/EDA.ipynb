## Problem Statement 

Late deliveries negatively impact customer satisfaction, increase operational costs, and affect brand reliability. The objective is to predict whether an order is likely to be delivered late at the time of order placement so that proactive actions can be taken.

Late delivery prediction is an operational decision problem, requiring order-level, real-time risk assessment rather than long-term demand forecasting.
